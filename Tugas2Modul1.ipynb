```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')

df.head()
```

```python
x = df['bmi'].values
y = df['charges'].values
```

```python
x = (x - x.mean()) / x.std()
y = (y - y.mean()) / y.std()
```

```python
w = 0.0
b = 0.0
learning_rate = 0.001
epochs = 1000

n = len(x)
```

```python
for epoch in range(epochs):
    for i in range(n):
        xi = x[i]
        yi = y[i]

        # prediksi
        y_pred = w * xi + b

        # error
        error = y_pred - yi

        # gradient
        dw = error * xi
        db = error

        # update
        w = w - learning_rate * dw
        b = b - learning_rate * db

    # monitoring loss
    if epoch % 100 == 0:
        loss = np.mean((w * x + b - y) ** 2)
        print(f"Epoch {epoch}, Loss: {loss}")
```

```python
print("Training selesai!")
print("w =", w)
print("b =", b)
```

```python
plt.scatter(x, y, label='Data')
plt.plot(x, w * x + b, label='Regresi')
plt.xlabel('BMI (normalized)')
plt.ylabel('Charges (normalized)')
plt.title('Linear Regression dengan SGD')
plt.legend()
plt.show()
```
